Поднимем контейнеры сначала для обычного редиса, а потом редис-кластер, описанный в папке redis_cluster.

Убедимся, что все контейнеры запущены

image.png

Теперь сравним скорость заполнения таблиц и скорость получения данных. Воспользуемся сриптом, который сначала заполняет данные, а потом получает их, замеряя время каждой операции.

В первом случае подключимся к кластеру, подключившись к порту 6376, на котором у нас хост
db = redis.StrictRedis(host='localhost', port=6376)

Во втором случае подключимся к обычному редису, через порт 6379
db = redis.StrictRedis(host='localhost', port=6379)

image.png

Видно, что время операций на обычном клиенте меньше, чем на кластере. Зато кластер повышает надежность данных
